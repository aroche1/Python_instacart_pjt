# Grouping Data with pandas

#### Coding data in pandas:
##### 1st. Split data into groups based on given data
##### 2nd. Apply function to these groups
##### 3rd. Combine results into dataframe or alternative structure or create a new column in the current dataframe

```python
# Split data into groups base on given data using 'groupby()' function
# This function groups a datframe by a given column
# Use it in 'product name' column in the 'df' (subset) dataframe

df.groupby('product_name')
```

# Aggregateing Data with 'agg()'

#### Aggregation function ctreate summaries of selected columns and store these summary values in new columns

#### These summaries usually take the form of descriptive statistics that can help standardize the data, for example, to be used in predictive modelling (reducing amount of variations in data)

## Performing a Single Aggregation

#### Aggregtin in Python is done by using 'agg()' function.

```python
# Get descriptive statistic fot hte 'order_number' column
# Calculate meand of 'order_number' column grouped by the 'apartment_id' column, (...)
# to compare the number of orders accross each Instacart departemnt

# 1st Split the data unto groupes bases on 'department_id'
# 2nd Apply 'agg()' function to each group to obtain the mean values for the 'order_number' column

df.groupby('department_id').agg({'order_number': ['mean']})
```

#### How to apply statistical test to proving significant difference between means of different groups (like using a t-test)?

```python
# Some aggregation can be ussed without using the 'agg()' function, and directly use a function especifically made to aggregate data in the way you may need, like 'mean()'

df.groupby('department_id')['order_number'].mean()
```

#### The syntax differs in that using 'agg()' function, the column to aggregate must be betweetn the function's parentheses;
#### whereas using 'mean()' function, or any other aggregation function, the columns has to be INDEXED with square brackets (square brackets have no other role in Python but to index)

## Performing Multiple Aggregations

#### It comes to using more argumnets to produce multiple statistics

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

# Aggregating Data with trnasform ()

#### After data has been aggregated, create flags for that data
#### Create a loyalty flag column in 'ords_prods_merge' dataframe
#### By using the 'transform()' fucntion, create a new column containing maximum frequency of the 'order_number' column, and then using 'loc()' function, create a second column to flag suctomers as 'loyal' or not

#### Criteria to create the flag:

##### If the maximum orders the customer has placed is over 40 orders, then 'Loyal customer'
##### If the maximum orders the user hasmade is over 10 but less than or equal to 40, then 'Regular customer'
##### If the maximum orders the user has made is less than or equal to 10, then 'New Customer'

#### Maping the task:
##### 1st. split the data into groups based on the 'user_id' column
##### 2nd. Apply the trnasform()' function on the 'order_number' column to generate the maximum orders for each user
##### 3rd. Create a new column, 'max_order', to place the results of aggregtion

#### Once this process is complete, use 'max_order' column to create another new colum tha assigns loyalty flag to each customer accordint to the criteria (via 'loc()' function).

```python
# Create 'max_order' column (all the three steps can be included in a single line of code)

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

##### First, a new column called “max_order” is created, which will store the maximum order number for each user (step 3)
##### Then, the 'ords-prods-merge' dataframe is grouped by the 'user_id' column (stpe1)
##### Finally, the 'trnasform()' function is applied on the on the 'order_number' colum with the np.max argument (step 2)

#### The funtion 'max()' from the NumPy (np) library is used to find the max value within a column
#### In this case, is being used within the 'trnsform()' function telling Python to:
#### "Trnasform the 'order_number' column by applying the 'max()' function from the NumPy" library"

# Deriving Columns with loc()

#### Time to set flag with column ready to go; assign a "loyalty" label to user ID based on it's corresponding max order value


```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order']>10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <=10, 'loyalty_flag'] = 'New customer'
```
