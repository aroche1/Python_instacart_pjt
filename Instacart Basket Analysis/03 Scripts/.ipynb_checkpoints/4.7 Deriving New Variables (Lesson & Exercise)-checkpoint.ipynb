```python
# User-Defined Function

def price_label(row):

  if row['prices'] <= 5: # If the value in the 'prices' column within the given row is less than or equal to 5, then ('then' is represented by colon)
    return 'Low-range product' # return the string 'Low-range product'.
  elif (row['prices'] > 5) and (row['prices'] <= 15): # Or else, if the value within the 'prices' columns within the given row is greater than 5 and less than or equal to 15, then 
    return 'Mid-range product' # return the string 'Mid-range product'.
  elif row['prices'] > 15: # Or else, if the value of the 'prices' column within the given row is greater than 15, then
    return 'High range' # return the string 'High-range product'.
  else: return 'Not enough data' # Or else, return the string 'Not enough data'
```

```python
# Now that the label has been created, we need to create a column where the output can be represented

df['price_range'] = df.apply(price_label, axis=1)

# "df['price_range']" is to create the column 'price_range' within the df (subset) dataframe
# Then, 'apply' the 'price_label' function on 'axis=1' that means that the function will be applied to all rows within the dataframe ('axis=0' would refer to columns in the dataframe) )
```

# If-Statements with the loc() Function

##### By using 'loc()', a conditional logic of an if-statement can be applied to a function without explicitly creating an 'if-else' function

##### Create conditions similar to the previous method
##### These conditions must be entered in the separate cells (?)

##### The 'loc()' function locates a particular column in the dataframe it's been assigned to.
##### A logical operatior is being added to the function (smaller than, larget than, equal to, etc.), which will act as a condition 
##### There's no explici 'if', instead it's been implied

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

###### If the values in 'prices' column of the 'df' dataframe are greater than '15', (after the come 'then' is implied) here a new column called 'price_range_loc' is being set equal to the string 'High-range product'

##### Two impied halves:
##### if = df.loc[df['prices'] > 15,
##### then = 'price_range_locs'] = 'High-range product'

###### from left to right
##### the 'loc()' function is being called on the 'df' (subset) dataframe.
##### Within the backets, the values in 'prices' column of the 'df' dataframe are being compared to a value of '15' using the '>' operator
##### The key is the 'coma' as it seperated the 'if' from 'then'

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

##### The same logic is applied, only this time I have two different arguments: Greater than '5' but less or equal to '15'
##### Combining these two conditons with '&' symbol and both conditions also separated/delimited by parentheses

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

##### If the 'prices' column of the 'df' dataframe is lower or equal to '5',  then a new column called 'price_range_loc' is set to equal the string 'low-range product'

```python
# Count the number of products within each label with 'calue_counts()' from the new 'price_range_loc' column

df['price_range_loc'].value_counts(dropna = False)
```

##### The results are the same than using the user-defined function
### Which method is better?
##### 1st. No 'warning message' with the result; Python thinks, at least, this the 'loc()' method is better
##### 2nd. The 'loc()' function runs faster because this fucntion applied the conditional filters before searching through the dataframe, 
##### while the user-defined function searched through the entire dataframe *****get more information on this****
##### Then, determined to set the filters (axis = 1)

```python
# Apply code to the entire (maind) dataframe

ords_prods_merged.loc[ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
ords_prods_merged.loc[(ords_prods_merged['prices'] <= 15) & (ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
ords_prods_merged.loc[ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

# If-Statements with For-Loops

##### 'For-loops' are loops to run a code multiple times during

```python
# Create a new column in 'ords_prods_merge' dataframe that summarize how busy each day od the week is.
# 1st, on which day most orders take place? What's the frequency of the 'orders_day_od_the_week' column?

ords_prods_merge['order_day_of_week'].value_counts(dropna = False) 
```

```python
# Create a column called ' busiest_day' that list the following values: 'Busiest Day', 'Least Busy', and 'Regular Busy'.
# Code a loop that runs through every row of 'order_day_of_week' column, compare values according to the set standard, and assign corresponding value.
# Codes to each day of the week:
# 0 = Saturday
# 1 = Sunday
# 2 = Monday
# 3 = Tuesday
# 4 = Wednesday 
# 5 = Thursday
# 6 = Friday

result = []

for value in ords_prods_merged["order_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Check 'result' variable

result
```

```python
# Combine 'result' new variable with 'ords_prods_merge' dataframe

ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

# Task 4.7: Deriving New Variables

```python
# 2.- Suppose the clients have changed their minds with the labels in 'busiest_day', and instead want 'Busiest day' (value),
# they want 'Busiest days' (value/plural) as label to correspond with the two busiest days of the week instead of one.
# Also, create the label for the two slowest days 
# Create a new column for these values
```

```python
# Creating for loop to list days of the week

result_2 = []

for value in ords_prods_merged['order_day_of_week']:
    if value in [0,1]:
        result_2.append('Busiest days')
    elif value in [3,4]:
        result_2.append('Slowest days')
    else:
        result_2.append('Intermediately busy days')
```

```python
# Check 'result_2'

result_2
```

```python
# Deriving 'busiest_days' column

ords_prods_merged['busiest_days'] = result_2
```

```python
# 3.- Check values in this new column for accuracy

ords_prods_merged['busiest_days'].value_counts(dropna = False)

```

```python
# 4.- Identify the busiest hours of the day, rathen than by hour and give the following labels: 
# 'Most orders', 'Average orders', and 'Fewest orders' under new calumn called 'busiest_period_of_day'
# Create a for loo to list the busiest hours of the day

result_3 = []

for value in ords_prods_merged['order_hour_of_day']:
  if value>=9 and value<=16:
    result_3.append("Most orders")
  elif value<=6:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
# Deriving 'busiest_period_of_day' column

ords_prods_merged['busiest_period_of_day'] = result_3
```

```python
# 5.- Check values in 'busiest_period_of_day' new column for accuracy

ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
