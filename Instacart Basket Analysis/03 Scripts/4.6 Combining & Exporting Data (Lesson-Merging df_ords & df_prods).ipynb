```python
# Merging 'df_ords' and 'df_ords_prior' using 'order_id' as key

df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
df_merged_large
```

```python
df_merged_large['_merge'].value_counts()

# Having all of the values as if they were in both dataframes is missleading, because of the way the dataframes weere merged: Inner Join
# The Inner Join will only match values in both dataframes, so this doesn't mean we have a full match
```

```python
# How to check whether I have a full match or not? 
# By specifying 'outer' in the 'how' argument like this: ', how = outer)'\
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', how = 'outer', indicator = True)
```

```python
df_merged_large['_merge'].value_counts()
```
