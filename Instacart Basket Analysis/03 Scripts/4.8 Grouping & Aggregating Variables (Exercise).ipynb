## 4.- Create loyalty flag for existing customers using the 'transform()' and 'loc()' functions.

#### Criteria to create the flag:

##### If the maximum orders the customer has placed is over 40 orders, then 'Loyal customer'
##### If the maximum orders the user hasmade is over 10 but less than or equal to 40, then 'Regular customer'
##### If the maximum orders the user has made is less than or equal to 10, then 'New Customer'

#### Once this process is complete, use 'max_order' column to create another new colum tha assigns loyalty flag to each customer accordint to the criteria (via 'loc()' function).

### Aggregating Data with trnasform()

```python
# Create 'max_order' column (all the three steps can be included in a single line of code)

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the newly created 'max_order' column

ords_prods_merge['max_order']
```

```python
# Check 'ords_prods_merge' dataframe

ords_prods_merge
```

#### A new colum called 'max_order' has been created, indicating the number of orders each customer ID has placed.

### Deriving Columns with loc()

#### Assign a loyalty flag to each customer ID based on corresponding 'max_column' value

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check flags for each type of customers

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## 5.- Marketing team wants to know whether there's a difference between the spending habits of the three identified type of customers. ##  6.- Determine the prices if products purchased by loyal customers differ from the other two types of customers, by creating an 'average_expenditure' and flagging each customer according to their expenditure behaviour

```python
ords_prods_merge['average_expenditure'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check 'average_expenditure' column

ords_prods_merge['average_expenditure']
```

### Determine Whther products purchased by loyal customers differ from those purchased by the other two types or customers by creating flags

```python
# flag for 'High Spender'
ords_prods_merge.loc[(ords_prods_merge['average_expenditure'] >= 10), 'expenditure_flag'] = 'High Spender'
```

```python
# flag for 'Low Spender'
ords_prods_merge.loc[(ords_prods_merge['average_expenditure'] < 10), 'expenditure_flag'] = 'Low Spender'
```

```python
# Check number of flags based on customers expenditure behaviour 

ords_prods_merge['expenditure_flag'].value_counts(dropna = False)
```

## 7.- In order to send relevant notification to (customer) users within the app, the Instacart team wants to determine frequent vs/ non-frequent customers. Create an order frequency flag that marks the regularity of user's ordering behaviour according to median in the 'days_since_prior_order column. The flag criteria is as follows

### -If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
### - If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
### - If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Creating 'purchase_frequency' column

ords_prods_merge['purchase_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check 'purchase_frequency' column
ords_prods_merge['purchase_frequency']
```

```python
ords_prods_merge.shape
```

```python
# Flag for 'Non-Frequent Customer'

ords_prods_merge.loc[(ords_prods_merge['purchase_frequency'] > 20), 'frequency_flag'] = 'Non-Frequent Cusotmer'
```

```python
# Flag for 'Regular Customer'

ords_prods_merge.loc[(ords_prods_merge['purchase_frequency'] > 10) & (ords_prods_merge['purchase_frequency'] <= 20), 'frequency_flag'] = 'Regular Customer'
```

```python
# Flag for ' Frequent Customer'

ords_prods_merge.loc[(ords_prods_merge['purchase_frequency'] <= 10), 'Frequency_flag'] = 'Frequent Customer'
```

```python
# Check frequency flags based on customers' behaviour

ords_prods_merge['expenditure_flag'].value_counts(dropna = False)
```
