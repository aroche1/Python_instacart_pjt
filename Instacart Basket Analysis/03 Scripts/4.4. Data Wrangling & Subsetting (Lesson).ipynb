# 05. Subsetting

```python
# Subsetting referes to creating a smaller data set from a whole data set, based on a particular filter
# What if I want to search for other prodcut in the same department? 
# Create a subset for the 'df_prods' dataframe that only contains data from the snack department:
df_snacks = df_prods[df_prods['department_id']==19] 
# within the brackets, requesting Python to look into the df_prods dataframe and find a column called 'department_id', and withon that column, look for the value '19'
# the 'object['element']' syntax is called indexing, which refers to retreiving some element from an object (the dataframe in this case, or a lsit, or otherwise).
# Bu using the index operator '[]', python will look for the element in the brackets within the object to the left of the brackets (looking for a 'department_id value of '19' within the 'df_prods' dataframe)
```

```python
df_snacks
```

```python
# breaking down the 'df_snacks' code.
# Python runs every single row and returns all the rows either with a 'True' ir '19' was found in a row or 'False' it it wasn't
df_prods['department_id']==19
```

```python
#adding up another instance: df_prods[]

df_prods[df_prods['department_id']==19]

# This code will ask Python to map the true/false procedure in the previous output onto the columns within the 'df_prods' dataframe. THe result is a lsit of those values.
```

```python
# Verifying the new 'df_snacks' dataframe was created
df_snacks.head()
```

```python
# There's a simpler way to achieve the same results:
# Using the 'loc' function (that's similar to 'iloc', which lookd for an integer position), which looks for a column name.
# Use this fucntion to look for instances where the value of 'department_id' is '19' within the df_prods database:
df_snack_2 = df_prods.loc[df_prods['department_id'] == 19]
```

```python
df_snack_2
```

```python
# or instead of using double '=' sign, better tell 'loc' function to look into a list: isin([19])
# With this fucntion i can look for multiple 'department_id' at the same time (isin([17, 18, 19)]))

df_snacks_3 = df_prods.loc[df_prods['department_id'].isin([19])]
```

```python
df_snacks_3
```
