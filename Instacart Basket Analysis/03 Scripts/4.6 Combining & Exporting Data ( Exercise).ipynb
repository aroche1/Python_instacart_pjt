# 03. Merging 'df_prods' & 'orders_products_combined' Dataframes

```python
# 5.- Determine a suitable way to combine 'orders_products_combined' dataframe with the 'products_checked.csv' data set

# 'df_prods' has 49672 records
# 'orders_products_combined' has 32434489 records
# Merging dataframes is the best method to combine dataframes
# As we are only working with full merge rates, we can just proced to make an Inner Join

df_ords_prods_merged = orders_products_combined.merge(df_prods, on = 'product_id', indicator = True)
```
